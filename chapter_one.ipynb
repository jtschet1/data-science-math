Graphing a linear function with Sympy

```python
from sympy import *
```

```python
x = symbols('x')
f = 2*x + 1
plot(f)
```

Note that the plot above is correct, but is distorted to portray the function at a 45 degree angle, The actual plot is a steep line because the slope is 2. To preserve this, we can set the aspect ratio to be 1:1, the ylim to be the axis_range, and the second argument which is the independent variable range to the axis_range.

```python
x = symbols('x')
f = 2*x + 1

# Define the range for both axes
axis_range = (-500, 500)

# Plot with equal aspect ratio and same limits for both axes
plot(f, (x, axis_range[0], axis_range[1]), ylim=axis_range, aspect_ratio=(1, 1))
```

Graphing exponential function with Sympy

```python
x = symbols('x')
f = x**2
plot(f)
```

Same with linear, we can adjust formatting

```python
x = symbols('x')
f = x**2

# Define the range for both axes
x_range = (-10, 10)
y_range = (0, 10)

# Plot with equal aspect ratio and same limits for both axes
plot(f, (x, x_range[0], x_range[1]), ylim=y_range, aspect_ratio=(1, 1))
```

Graphing multiple variable functions with Sympy

```python
# from sympy import * already imported earlier
from sympy.plotting import plot3d
x, y = symbols('x y')
f = x + y
plot3d(f)
```

Performing a Summation in Python

```python
# without a list. Summation of numbers from 1 to 5 multiplied by 2
summation = sum(2*i for i in range(1, 6))
print(summation)

# with a list. Summation of numbers in list x multiplied by 10
x = [1, 4, 6, 2]
n = len(x)

summation = sum(10*x[i] for i in range(0, n))
print(summation)
```

Summations using Sympy

```python
# from sympy import *
i,n = symbols('i n')
# iterate each element i from 1 to n,
# then multiply and sum
summation = Sum(2*i,(i,1,n))
# specify n as 5,
# iterating the numbers 1 through 5
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit()) # 30

#the above code calculates the summation of numbers from 1 to 5 multiplied by 2, identical to this code:
summation = sum(2*i for i in range(1, 6))
print(summation)
```

Exponents:
Product rule and Power rule. When multiplying two exponents with the same base, add the exponents. When dividing two exponents with the same base, subtract the exponents. When raising an exponent to another exponent, multiply the exponents.

Multiplying Example:
$2^3 * 2^4 = 2^{3+4} = 2^7$

Dividing Example:
$2^5 / 2^3 = 2^{5-3} = 2^2$

Raising to an Exponent Example:
$(2^3)^4 = 2^{3*4} = 2^{12}$
