Exponents:
Product rule and Power rule. When multiplying two exponents with the same base, add the exponents. When dividing two exponents with the same base, subtract the exponents. When raising an exponent to another exponent, multiply the exponents.

Multiplying Example:
$2^3 * 2^4 = 2^{3+4} = 2^7$

Dividing Example:
$2^5 / 2^3 = 2^{5-3} = 2^2$

Raising to an Exponent Example:
$(2^3)^4 = 2^{3*4} = 2^{12}$

Raising to a fraction Example:
$2^{3/2} = \sqrt{2^3} = \sqrt{8}$

Logarithm and Exponent Properties:

| Operator       | Exponent Property          | Logarithm Property                |
|----------------|----------------------------|-----------------------------------|
| Multiplication | $x^m \cdot x^n = x^{m+n}$  | $\log(a \cdot b) = \log(a) + \log(b)$ |
| Division       | $\frac{x^m}{x^n} = x^{m-n}$| $\log\left(\frac{a}{b}\right) = \log(a) - \log(b)$ |
| Exponentiation | $(x^m)^n = x^{mn}$         | $\log(a^n) = n \cdot \log(a)$     |
| Zero Exponent  | $x^0 = 1$                  | $\log(1) = 0$                     |
| Inverse        | $x^{-1} = \frac{1}{x}$     | $\log(x^{-1}) = \log\left(\frac{1}{x}\right) = -\log(x)$ |

Euler's Number: a special number, one property of which is that its exponential function is its own derivative, meaning that the slope of the function at any point is equal to the value of the function at that point.

Compound interest formula: $A = P(1 + r/n)^{nt}$

Where: variables are A; balance, P; principal, r; rate, n; number of times compounded, t; time

````python
#formula for calculating compound interest
# variables are A; balance, P; principal, r; rate, n; number of times compounded, t; time
# the formula is A = P * (1 + r/n)**(n*t)
# from sympy import *
def calc_interest(principal, rate, num_compounds, time):
    return principal * (1 + rate/num_compounds) ** (num_compounds*time)

#if a principal of $100 is invested at a 20% annual interest, compounded monthly, for 2 years, what is the balance?
p = 100
r = .20
n = 12
t = 2
a = calc_interest(p, r, n, t)
print(a)
````

You can change the value of n to compound the interest more or less frequently. The more frequently you compound the interest, the more money you will have at the end of the period, if taken to the extreme and make the periods of compounding infinitesimally small, you will have the formula for continuous compounding. This is where Euler's number comes in.

 The formula for continuous compounding is: $A = P \cdot e^{rt}$

Where: variables are A; balance, P; principal, r; rate, t; time, and e (technically a constant) is Euler's number.


````python
from math import exp
p = 100 # principal, starting amount
r = .20 # interest rate, by year
t = 2.0 # time, number of years
a = p * exp(r*t)
print(a) 
````

Where do we derive Euler's number from? It is the limit of $(1 + \frac{1}{n})^n$ as n approaches infinity. This is a special case of the compound interest formula where the interest is compounded an infinite number of times. For any value of n, the formula will give you a number close to Euler's number, but as n approaches infinity, the number will converge to Euler's number. 

The equation for this would be $e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n$ where e is Euler's number, n is the number of times compounded, and the limit is taken as n approaches infinity.

Derivatives: The a derivative is the slope of a function. Used often for gradient descent. This helps with linear regression, logistic regression, and neural networks.

Examining the function $f(x) = x^2$ we can see that the slope of the function where $x = 2$ is 4. This is because the derivative of $x^2$ is $2x$. Therefore, the slope of the function at $x = 2$ is $2(2) = 4$.

````python
# we can graph a quadratic equation and the tangent line at x = 2 to show the slope of the tangent line
from sympy import *
x = symbols('x')
f = x**2
f_prime = f.diff(x)
# tangent =    [        slope        ] [  shift y ]    [   shift   x        ]
tangent_line = f_prime.subs(x, 2) * x + f.subs(x, 2) - f_prime.subs(x, 2) * 2
# y - y1 = m(x - x1)
# y = m(x - x1) + y1
# y =               m*x                 + y1                     - m*x1
#label the tangent line and the quadratic equation and plot them both
plot(f, f_prime, tangent_line, (x, -10, 10), legend=True)
````

The derivative_x function is the same as the formula for a slope between two points $m = \frac{y_2 - y_1}{x_2 - x_1}$

````python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m

def f(x):
    return x**2

slope_at_2 = derivative_x(f, 2, 0.00001)

print(slope_at_2)
````

Partial Derivatives: These are derivatives of functions that have multiple input variables. Think of it this way. Rather than finding the slope on a one-dimensional function,
we have slopes with respect to multiple variables in several directions. The function $f(x, y) = x^2 + y^2$ has two partial derivatives, $\frac{\partial f}{\partial x}$ and $\frac{\partial f}{\partial y}$.

For example we have the function $f(x, y) = 2x^3 + 3y^3$. The partial derivative with respect to x is $\frac{\partial f}{\partial x} = 6x^2$ and the partial derivative with respect to y is $\frac{\partial f}{\partial y} = 9y^2$.

We can calculate partial derivatives with Sympy

````python
from sympy import *
from sympy.plotting import plot3d
# Declare x and y to SymPy
x,y = symbols('x y')
# Now just use Python syntax to declare function
f = 2*x**3 + 3*y**3
# Calculate the partial derivatives for x and y
dx_f = diff(f, x)

dy_f = diff(f, y)
print(dx_f) # prints 6*x**2
print(dy_f) # prints 9*y**2
# plot the function
plot3d(f)
````

Using limits to calculate derivatives is useful because it infinetly decreases our "step size" that we talked about earlier to get as close as we can to the slope of a single point. In math that function to approach 0 to calculate a slope would be $\lim_{s \to 0} \frac{f(x + s) - f(x)}{(x + s) - x}$

For f(x) = x^2, that looks like $\lim_{s \to 0} \frac{(x + s)^2 - x^2}{(x + s) - x}$

For f(x) = x^2 where x = 2, that looks like $\lim_{s \to 0} \frac{(2 + s)^2 - 2^2}{(2 + s) - 2}$ = $\lim_{s \to 0} \frac{4 + 4s + s^2 - 4}{s}$ = $\lim_{s \to 0} \frac{4s + s^2}{s}$ = $\lim_{s \to 0} 4 + s$ = 4

We can use limits to calculate a slope with Sympy

````python
from sympy import *
# "x" and step size "s"
x, s = symbols('x s')
# declare function
f = x**2
# slope between two points with gap "s"
# substitute into rise-over-run formula
slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)
# substitute 2 for x
slope_2 = slope_f.subs(x, 2)
# calculate slope at x = 2
# infinitely approach step size _s_ to 0
result = limit(slope_2, s, 0)
print(result) # 4

#note if do not substitute x = 2, the result will be 2*x + s
x, s = symbols('x s')

f = x**2

slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)

result = limit(slope_f, s, 0)

print(result) # 2*x
````

Chain Rule: The chain rule is useful for building neural networks which says that for a given function y (with input variable x)
composed into another function z (with input variable y), we can find the derivative of z with respect to x by multiplying the two respective derivatives together: This is expressed as $\frac{dz}{dx} = \frac{dz}{dy}*\frac{dy}{dx}$   For example, if we have the function $f(x) = (x^2 + 1)^3$, the derivative of the outer function is $3(x^2 + 1)^2$ and the derivative of the inner function is $2x$. Therefore, the derivative of the composite function is $3(x^2 + 1)^2 * 2x$.

Example: 

$y = x^2 + 1$

$z = y^3 -2$

$z = (x^2 + 1)^3 - 2$

$\frac{dz}{dx} = 3(x^2 + 1)^2 * 2x$

$\frac{dz}{dx} = 6x(x^2 + 1)^2$

We can calculate the chain rule with Sympy

````python
from sympy import *
z = (x**2 + 1)**3 - 2
dz_dx = diff(z, x)
print(dz_dx)
````

Integrals: The integral is the oppposite of the derivative. It is the area under the curve of a function. This is useful for calculating the area under a curve, which is useful for calculating probabilities in statistics. There is an approach to learning integrals called Riemann sums, which is the sum of the areas of rectangles that approximate the area under a curve. The more rectangles you use, the more accurate the approximation. The integral is the limit of the Riemann sum as the number of rectangles approaches infinity.

For a curved line, which many functions are, this is tricky, because you can't simply count the triangles under the curve. Instead, you have to use calculus to find the area under the curve. What we can do is use a method called Riemann sums, which is the sum of the areas of rectangles that approximate the area under a curve. The more rectangles you use, the more accurate the approximation. The integral is the limit of the Riemann sum as the number of rectangles approaches infinity.

````python
#finding the area under a curve for an easy linear function like 2x + 1
# plot a graph with the area under the curve shaded
def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0
    
    for i in range(1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)
    return total_sum * delta_x

def my_function(x):
    return x**2 + 1

area = approximate_integral(a=0, b=1, n=4, f=my_function)
area2 = approximate_integral(a=0, b=1, n=100, f=my_function)
area3 = approximate_integral(a=0, b=1, n=1000, f=my_function)
area4 = approximate_integral(a=0, b=1, n=10000, f=my_function)
# in this case, n is the number of rectangles under the curve, the higher the number, the more accurate the area
print(area)
print(area2)
print(area3)
print(area4)
````

````python
# we can do this more precisely with sympy
from sympy import *
# Declare 'x' to SymPy
x = symbols('x')
# Now just use Python syntax to declare function
f = x**2 + 1
# Calculate the integral of the function with respect to x
# for the area between x = 0 and 1
area = integrate(f, (x, 0, 1))
print(area) # prints 4/3
# notice how sympy supports rational numbers, the result is 4/3 rather than a repeating decimal
````
